```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R1 = 0.1
Vimax = 6
Rmax = 9
Vi = np.linspace(0,Vimax,21)

lgnd = []

for r in range(1,Rmax):
    i = Vi/(R1+r)
    p2 = r * i**2
    plt.plot(Vi,p2)
    lgnd.append(f'{r} Ohm')
plt.title('Load power (P2) as function of input voltage (Vi)')
plt.xlabel('Vi')
plt.ylabel('P2 [W]')
plt.ylim(0,5)
plt.legend(lgnd)
plt.grid(True)
plt.show()
```

## Power as function of load resistance with input voltage as parameter

$ P(r) = {V_i}^2\frac{r}{(R+r)^2} $

The maximum of this function is when R=r (maximum transfer of power)
If:  
$ R=0.1\Omega, Vi=1V $ \
$ P_{max} = 2.5 {V_i}^2 = 2.5  $

One has to be careful how he or she uses the rheostat.
Even though the specified maximum power of a 9 Ohm rheostat is 100W, it applies only when the full resistance participates in the circuit.
If a smaller fraction of the rheostat is part of the electical circuit the max power will be proportionally smaller.
If only 2 Ohms are used, the maximum power will be 2/9 * 100W = 22.2W

For Vi=5V: 
P = 2 / (2.1)**2 * 25

It should still survive the following conditions (if specified max power was 100W)
P = 22.2W
R = 2 Ohm
I = sqrt(P/R) = sqrt(22.2/2) = sqrt(11) = 3.33A
V = P/I = 6.66V
